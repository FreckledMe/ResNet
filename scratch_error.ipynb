```python
class ResNet(keras.Model):
    def __init__(self):
        super(ResNet,self).__init__()

        # block 1
        self.block1_1 = Conv2D(64,kernel_size=(7,7),strides=(2,2),padding='same',activation='relu')
        self.block1_2 = MaxPooling2D(pool_size=(2,2),padding='same')

        # block 2
        self.block2_1 = Conv2D(64,kernel_size=(3,3),padding='same')
        self.block2_2 = BatchNormalization()
        self.block2_3 = ReLU()

        self.block2_4 = Conv2D(64,kernel_size=(3,3),padding='same')
        self.block2_5 = BatchNormalization()
        self.block2_6 = ReLU()

        self.block2_7 = Conv2D(64,kernel_size=(3,3),padding='same')
        self.block2_8 = BatchNormalization()
        self.block2_9 = ReLU()

        self.block2_10 = Conv2D(64,kernel_size=(3,3),padding='same')
        self.block2_11 = BatchNormalization()
        self.block2_12 = ReLU()

        self.block2_13 = Conv2D(64,kernel_size=(3,3),padding='same')
        self.block2_14 = BatchNormalization()
        self.block2_15 = ReLU()

        self.block2_16 = Conv2D(64,kernel_size=(3,3),padding='same')
        self.block2_17 = BatchNormalization()
        self.block2_18 = ReLU()

        # block 3
        self.block3_1 = Conv2D(128,kernel_size=(3,3),strides=(2,2),padding='same')
        self.block3_2 = BatchNormalization()
        self.block3_3 = ReLU()

        self.block3_4 = Conv2D(128,kernel_size=(3,3),padding='same')
        self.block3_5 = BatchNormalization()
        self.block3_6 = ReLU()

        self.block3_7 = Conv2D(128,kernel_size=(3,3),padding='same')
        self.block3_8 = BatchNormalization()
        self.block3_9 = ReLU()

        self.block3_10 = Conv2D(128,kernel_size=(3,3),padding='same')
        self.block3_11 = BatchNormalization()
        self.block3_12 = ReLU()

        self.block3_13 = Conv2D(128,kernel_size=(3,3),padding='same')
        self.block3_14 = BatchNormalization()
        self.block3_15 = ReLU()

        self.block3_16 = Conv2D(128,kernel_size=(3,3),padding='same')
        self.block3_17 = BatchNormalization()
        self.block3_18 = ReLU()

        self.block3_19 = Conv2D(128,kernel_size=(3,3),padding='same')
        self.block3_20 = BatchNormalization()
        self.block3_21 = ReLU()

        self.block3_22 = Conv2D(128,kernel_size=(3,3),padding='same')
        self.block3_23 = BatchNormalization()
        self.block3_24 = ReLU()

        # block 4
        self.block4_1 = Conv2D(256,kernel_size=(3,3),strides=(2,2),padding='same')
        self.block4_2 = BatchNormalization()
        self.block4_3 = ReLU()

        self.block4_4 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block4_5 = BatchNormalization()
        self.block4_6 = ReLU()

        self.block4_7 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block4_8 = BatchNormalization()
        self.block4_9 = ReLU()

        self.block4_10 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block4_11 = BatchNormalization()
        self.block4_12 = ReLU()

        self.block4_13 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block4_14 = BatchNormalization()
        self.block4_15 = ReLU()

        self.block4_16 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block4_17 = BatchNormalization()
        self.block4_18 = ReLU()

        self.block4_19 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block4_20 = BatchNormalization()
        self.block4_21 = ReLU()

        self.block4_22 = Conv2D(256,kernel_size=(3,3),strides=(2,2),padding='same')
        self.block4_23 = BatchNormalization()
        self.block4_24 = ReLU()

        self.block4_25 = Conv2D(256,kernel_size=(3,3),strides=(2,2),padding='same')
        self.block4_26 = BatchNormalization()
        self.block4_27 = ReLU()

        self.block4_28 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block4_29 = BatchNormalization()
        self.block4_30 = ReLU()

        self.block4_31 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block4_32 = BatchNormalization()
        self.block4_33 = ReLU()

        self.block4_34 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block4_35 = BatchNormalization()
        self.block4_36 = ReLU()

        # block 5
        self.block5_1 = Conv2D(256,kernel_size=(3,3),strides=(2,2),padding='same')
        self.block5_2 = BatchNormalization()
        self.block5_3 = ReLU()

        self.block5_4 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block5_5 = BatchNormalization()
        self.block5_6 = ReLU()

        self.block5_7 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block5_8 = BatchNormalization()
        self.block5_9 = ReLU()

        self.block5_10 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block5_11 = BatchNormalization()
        self.block5_12 = ReLU()

        self.block5_13 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block5_14 = BatchNormalization()
        self.block5_15 = ReLU()

        self.block5_16 = Conv2D(256,kernel_size=(3,3),padding='same')
        self.block5_17 = BatchNormalization()
        self.block5_18 = ReLU()

        # block 6
        self.avg_pool = AveragePooling2D(pool_size=(2,2),strides=(2,2))
        self.flatten = Flatten()

        self.fc = Dense(15,activation='softmax')

    def call(self,inputs):

        x = self.block1_1(inputs)
        x = self.block1_2(x)

        out_block1 = Add([inputs,x])

        x = self.block2_1(out_block1)
        x = self.block2_2(x)
        x = self.block2_3(x)
        x = self.block2_4(x)
        x = self.block2_5(x)
        x = self.block2_6(x)

        out_block2_1 = Add([out_block1,x])

        x = self.block2_7(out_block2_1)
        x = self.block2_8(x)
        x = self.block2_9(x)
        x = self.block2_10(x)
        x = self.block2_11(x)
        x = self.block2_12(x)

        out_block2_2 = Add([out_block2_1,x])

        x = self.block2_13(out_block2_2)
        x = self.block2_14(x)
        x = self.block2_15(x)
        x = self.block2_16(x)
        x = self.block2_17(x)
        x = self.block2_18(x)

        out_block2_3 = Add([out_block2_2,x])

        x = self.block3_1(out_block2_3)
        x = self.block3_2(x)
        x = self.block3_3(x)
        x = self.block3_4(x)
        x = self.block3_5(x)
        x = self.block3_6(x)

        out_block3_1 = Add([out_block2_3,x])

        x = self.block3_7(out_block3_1)
        x = self.block3_8(x)
        x = self.block3_9(x)
        x = self.block3_10(x)
        x = self.block3_11(x)
        x = self.block3_12(x)

        out_block3_2 = Add([out_block3_1,x])

        x = self.block3_13(out_block3_2)
        x = self.block3_14(x)
        x = self.block3_15(x)
        x = self.block3_16(x)
        x = self.block3_17(x)
        x = self.block3_18(x)

        out_block3_3 = Add([out_block3_2,x])

        x = self.block3_19(out_block3_2)
        x = self.block3_20(x)
        x = self.block3_21(x)
        x = self.block3_22(x)
        x = self.block3_23(x)
        x = self.block3_24(x)

        out_block3_4 = Add([out_block3_3,x])

        x = self.block4_1(out_block3_4)
        x = self.block4_2(x)
        x = self.block4_3(x)
        x = self.block4_4(x)
        x = self.block4_5(x)
        x = self.block4_6(x)

        out_block4_1 = Add([out_block3_4,x])

        x = self.block4_7(out_block3_4)
        x = self.block4_8(x)
        x = self.block4_9(x)
        x = self.block4_10(x)
        x = self.block4_11(x)
        x = self.block4_12(x)

        out_block4_2 = Add([out_block4_1,x])

        x = self.block4_13(out_block4_2)
        x = self.block4_14(x)
        x = self.block4_15(x)
        x = self.block4_16(x)
        x = self.block4_17(x)
        x = self.block4_18(x)
        
        out_block4_3 = Add([out_block4_2,x])

        x = self.block4_19(out_block4_3)
        x = self.block4_20(x)
        x = self.block4_21(x)
        x = self.block4_22(x)
        x = self.block4_23(x)
        x = self.block4_24(x)

        out_block4_4 = Add([out_block4_3,x])
        
        x = self.block4_25(out_block4_4)
        x = self.block4_26(x)
        x = self.block4_27(x)
        x = self.block4_28(x)
        x = self.block4_29(x)
        x = self.block4_30(x)

        out_block4_5 = Add([out_block4_4,x])

        x = self.block4_31(out_block4_5)
        x = self.block4_32(x)
        x = self.block4_33(x)
        x = self.block4_34(x)
        x = self.block4_35(x)
        x = self.block4_36(x)

        out_block4_6 = Add([out_block4_5,x]) 

        x = self.block5_1(out_block4_6)
        x = self.block5_2(x)
        x = self.block5_3(x)
        x = self.block5_4(x)
        x = self.block5_5(x)
        x = self.block5_6(x)

        out_block5_1 = Add([out_block4_6,x])

        x = self.block5_7(out_block5_1)
        x = self.block5_8(x)
        x = self.block5_9(x)
        x = self.block5_10(x)
        x = self.block5_11(x)
        x = self.block5_12(x)

        out_block5_2 = Add([out_block5_1,x])

        x = self.block5_13(out_block4_2)
        x = self.block5_14(x)
        x = self.block5_15(x)
        x = self.block5_16(x)
        x = self.block5_17(x)
        x = self.block5_18(x)

        out_block5_3 = Add([out_block5_2,x])

        x = self.avg_pool(out_block5_3)

        x = self.flatten(x)

        x = self.fc(x)

        return x
```
